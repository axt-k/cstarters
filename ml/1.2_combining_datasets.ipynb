```python
import pandas as pd
```

```python
c_start_manual_df = pd.read_csv("csv/C_starter_ML_gsheet.csv")
cs_df = pd.read_csv("csv/condensation_starter.csv")
gt_df = pd.read_csv("csv/ground_truth.csv")
```

```python
# Create a dictionary mapping values in 'Unnamed: 3' column to their row indices
unnamed_3_to_index = {}
for idx, value in enumerate(gt_df['Unnamed: 3']):
    if pd.notna(value):  # Only include non-NaN values
        unnamed_3_to_index[value] = idx

# Display the original dataframe and the mapping dictionary
print("Dictionary mapping 'Unnamed: 3' values to row indices:")

# Display the first few rows of the ground truth dataframe
print(gt_df)

# Create a new dataframe with accession and category from cs_df
accession_category_map = cs_df[['accession', 'category']].dropna(subset=['category']).drop_duplicates()

category_dict = dict(zip(accession_category_map['accession'], accession_category_map['category']))

accession_3OH_map = cs_df[['accession', '3OH']].dropna(subset=['3OH']).drop_duplicates()
# Create a dictionary for quick lookup
boh_dict = dict(zip(accession_3OH_map['accession'], accession_3OH_map['3OH']))

smiles_map = c_start_manual_df[['accession', 'Donor SMILES']].dropna(subset=['Donor SMILES']).drop_duplicates()
smiles_dict = dict(zip(smiles_map['accession'], smiles_map['Donor SMILES']))


for index, row in gt_df.iterrows():
    if gt_df.at[index, 'Unnamed: 3'] in category_dict:
        gt_df.at[index, 'Unnamed: 22'] = category_dict[gt_df.at[index, 'Unnamed: 3']]
    if gt_df.at[index, 'Unnamed: 3'] in boh_dict:
        gt_df.at[index, 'Unnamed: 21'] = category_dict[gt_df.at[index, 'Unnamed: 3']]
    if gt_df.at[index, 'Unnamed: 3'] in smiles_dict:
        gt_df.at[index, 'curated'] = smiles_dict[gt_df.at[index, 'Unnamed: 3']]
gt_df.to_csv('csv/ground_truth_curated.csv', index=False)
```
