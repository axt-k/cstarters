#### split csv into train and final holdout

```python
from sklearn.model_selection import train_test_split
import pandas as pd

df = pd.read_csv('csv/condensation_starter.csv')
df = df[df['hole'].notna()]
train, test = train_test_split(df, test_size=0.2, stratify=df['3OH'],random_state=1)

train.to_csv('csv/condensation_starter_train.csv', index=False)
test.to_csv('csv/condensation_starter_test.csv', index=False)
```

```python
import pandas as pd
df_tax = pd.read_csv('cs_data_withtax.csv')

df_c_train = pd.read_csv('csv/condensation_starter_train.csv')
df_c_test = pd.read_csv('csv/condensation_starter_test.csv')
df_c_ = pd.read_csv('csv/condensation_starter.csv')

tax_ref = {}
for i in range(len(df_tax)):
    phyl = df_tax.at[i,'full_tax']
    
    phyl_cat = 'other'
    if str(phyl) != 'nan':
        if 'Pseudomonas' in phyl:
            phyl_cat = 'pseudomonas'
        if 'Streptomyces' in phyl:
            phyl_cat = 'streptomyces'


    tax_ref[df_tax.at[i,'accession']] = phyl_cat


for i in range(len(df_c_train)):
    
    acc = df_c_train.at[i, 'accession']
    phyl_cat = tax_ref[acc]
    df_c_train.at[i,'tax_cat'] = phyl_cat


for i in range(len(df_c_test)):
    
    acc = df_c_test.at[i, 'accession']
    phyl_cat = tax_ref[acc]
    df_c_test.at[i,'tax_cat'] = phyl_cat


for i in range(len(df_c_)):
    
    acc = df_c_.at[i, 'accession']
    phyl_cat = tax_ref[acc]
    df_c_.at[i,'tax_cat'] = phyl_cat


df_c_train.to_csv('csv/condensation_starter_train.csv', index=False)
df_c_test.to_csv('csv/condensation_starter_test.csv', index=False)

df_c_.to_csv('csv/condensation_starter.csv', index=False)
df_c_
```
